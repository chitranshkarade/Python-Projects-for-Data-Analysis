# 8. Create co-relation map for visualization  

```python
corr_df = tracks.drop(['key','mode','explicit'], axis = 1).corr(method = 'pearson')
```

```python
plt.figure(figsize= (18,8))
```

```python
heatmap = sns.heatmap(corr_df,vmin=-1,
    vmax=1,
    cmap='inferno',
    center=0,
    annot=True,
    fmt='.1g',
    linewidths=3,
    )
heatmap.set_title("correlation heatmap Among Variable")
heatmap.set_xticklabels(heatmap.get_xticklabels(),rotation= 90)
```

```python
# Report:- here we can see energy increases with loudness , danceability increases with valence and many more 
```

# 9. Now we will take  0.4% data as a sample data & create regression plot using data  .

```python
sample_df = tracks.sample(int (0.004 * len(tracks)))
```

```python
sample_df
```

# Report:- 


# "loudness" vs "energy"

```python
plt.figure(figsize=(10,6))
sns.regplot(data = sample_df , y= "loudness" , x = "energy",color = "c").set(title = "loudness vs energy corelation  ")

```

# If the popularity increases when acousticness decreases

```python
plt.figure(figsize=(10,6))
sns.regplot(data = sample_df , y= "popularity" , x = "acousticness",color = "c").set(title = "Popularity vs acousticness corelation  ") 
```

Adding new column

```python
tracks['dates'] = tracks.index.get_level_values('release_date')
tracks.dates = pd.to_datetime(tracks.dates)
years = tracks.dates.dt.year
```

# Number of songs Created in each year. 

```python
sns.displot(years, discrete = True , aspect = 2, height = 5 , kind = 'hist').set(title = 'Number of songs per year')
```

# Duration of songs are increasing by years and then decreasing after 2010.

```python
total_dr = tracks.duration 
fig_dims = (18,7)
fig ,ax = plt.subplots(figsize = fig_dims)
fig = sns.barplot(x = years , y = total_dr , ax =ax , errwidth = False).set(title = 'year vs duration')
plt.xticks(rotation =90)
```
