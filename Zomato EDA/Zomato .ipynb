## Analysing each and every columns -

### Column - "rate"

```python
df["rate"].head()
```

```python
df["rate"].unique()
```

#### Rating we have provided is outof 5 so we will convert it into a float removing "/5" part of it then we will deal with "NEW" , "-" 

```python
def handling_rate(value):
    if (value== "NEW" or value == "-"):
        return np.nan
    else:
        value = str(value).split("/")
        value = value[0]
        return float(value)
df["rate"] = df["rate"].apply(handling_rate)
df["rate"].head()
```

```python
df.head()
```

```python
df = df.drop(["reviews_list"] , axis =1) # removing the reviews_list column 
```

```python
df.head()
```

### Fixing null values -

```python
df.rate.isnull().sum()
```

```python
df["rate"].fillna(df["rate"].mean(),inplace = True)
df["rate"].isnull().sum()
```

```python
df.isnull().sum()
```

```python
'''
fixing other null values  
1. location can't be fixed . Data insufiecient issue.
2. same goes with rest_type, cuisines, approx_cost(for two people)
therefore - 
''' 
```

```python
df.dropna(inplace = True)
```

```python
df.isnull().sum()
```

### Renaming columns -

```python
df.head()
```

```python
df.rename(columns = {"approx_cost(for two people)":"cost2","listed_in(type)":"types"},inplace = True )
df.head()
```

### Column - "location"

```python
df["location"].isnull().sum()
```

```python
df["location"].unique()  # everythings is right with location so moving to next section
```

### Column - "cost2" - 

```python
df["cost2"].unique()
```

```python
'''

Here data looks normal but when we closly observe values > 1000 has a comma in between therefore it wont be 
converted into float type

'''
```

```python
def handling_cost2(value):
    if "," in value:
        value = value.replace(",","")
        return float(value)
    else :
        return float(value)
df["cost2"] = df["cost2"].apply(handling_cost2)
df["cost2"].unique()
```

```python
df.head()
```

### Column - "cuisines"

```python
df["cuisines"].unique()
```

```python
df["cuisines"].value_counts()
```

```python
"""
After carefull observation of the values we can observe that their are many cuisines which has no repetation and has low 
value count which has now direct effect on the obseration so we have moved those value in the "other" section .

"""
```

```python
df1 = df["cuisines"].value_counts(ascending=False)
```

```python
cuisineslessthan100 = df1[df1<100]
```

```python
def handling_cuisines(value):
    if value in cuisineslessthan100:
        return "other"
    else :
        return value
df["cuisines"]=df["cuisines"].apply(handling_cuisines)
df["cuisines"].value_counts()
```

### Column - "location"

```python
df.head()
```

```python
df["location"].unique()
```

```python
df["location"].value_counts()
```

##### We have the same observation as "cuisines" column so we will handle it accordingly.

```python
location1 = df["location"].value_counts()
```

```python
locationlessthan500 = location1[location1<300]
```

```python
def handlinglocation(value):
    if value in locationlessthan500:
        return "others"
    else:
        return value
df["location"] = df["location"].apply(handlinglocation)
df["location"].value_counts()
```

### Column - "rest_type"

```python
df.head()
```

```python
df['rest_type'].value_counts()
```

```python
rest_type = df["rest_type"].value_counts(ascending = False)
```

```python
resttypelessthan1000 = rest_type[rest_type<1000]
```

```python
def handlingresttype(value):
    if value in resttypelessthan1000:
        return "others"
    else :
        return value
df["rest_type"] = df["rest_type"].apply(handlingresttype)
df["rest_type"].value_counts()
```
