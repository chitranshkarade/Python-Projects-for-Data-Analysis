## Handling columns one by one - 

```python
data.columns
```

### Column - "Name" 

```python
df = data 
```

```python
df["Name"].unique()
```

### Column - "pack_size"

```python
df["pack_size"].unique()
```

```python
df["pack_size"].value_counts()
```

```python
def handlingpacksize(value):
    value = value.split(" ")
    if value[2] =="of":
        return float(value[3])
    else :
        return float(value[2])
df["size"] = df["pack_size"].apply(handlingpacksize)
df["size"].unique
```

```python
df.head(10)
```

- We will also create a column name Type to have an understanding of what type size is given.

```python
def creatingcolumnsize(value):
    value = value.split(" ")
    return value[-1]
df["Type"] = df["pack_size"].apply(creatingcolumnsize)
df["Type"].unique()
```

### Column -"rating_count"

- We can observe it has 'No reviews' and "ratings" written which is not required.

```python
df["rating_count"].value_counts() 
```

```python
def handlingrating_count(value):
    value = value.split(" ")
    if value[0] == "No":
        return np.nan
    else :
        return int(value[0])
df["ratings"] = df["rating_count"].apply(handlingrating_count)
df["ratings"].unique()
```

```python
df.columns
```

```python
df.head()
```

### Column - "rating"

```python
df["rating"].unique()
```

- We need to convert the values in the Float type.

```python
def handlingrating(value):
    if value == "Nan":
        return np.nan
    else :
        return float(value)
df["Avg_rating"] = df["rating"].apply(handlingrating)
df["Avg_rating"].unique()
```

```python
df.head()
```

### Column - "price" -

```python
df["price"].value_counts()
```

```python
df.info()
```

```python
df['price'] = df['price'].apply(lambda x: x.replace('₹', '')) #remove the '₹' in the price column
df['price'] = df['price'].replace('No discounts', 0) #convert the "no discount" values to "O"
df['price'] = df['price'].astype('float')
```

```python
df.head()
```

### Column - "mrp"

```python
df['mrp'] = df['mrp'].apply(lambda x: x.replace('₹', ''))   #remove the '₹' in the price column
df['mrp'] = df['mrp'].apply(lambda x: x.replace('MRP', '')) #remove the "MRB" strnig in value
df['mrp'] = df['mrp'].astype('float')
```

```python
df.head()
```

```python
df.isnull().sum()
```

### Handling NAN value-

```python
df.fillna(0)
```

```python
df.isnull().sum()
```

```python
df.head()
```

```python
df.drop(["pack_size","rating_count","rating"],axis = 1,inplace =True)
```

```python
df.head()
```
